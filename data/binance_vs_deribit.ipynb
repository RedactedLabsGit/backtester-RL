```python
import pandas as pd

pd.options.plotting.backend = "plotly"
```

```python
binance_bvol = pd.read_csv("binance/ETHBVOLUSDT-2024-01>05.csv", index_col=0)
deribit_vol_index = pd.read_csv("deribit/ETH_volIndex.csv", index_col=0)
```

```python
binance_bvol.index = pd.to_datetime(binance_bvol.index, unit="s")
binance_bvol_1h = binance_bvol.resample("1H").last()

ax = pd.DataFrame({
    "deribit_vol_index": deribit_vol_index["vol_index"][:len(binance_bvol_1h["bvol"])]*100,
    "binance_bvol": binance_bvol_1h["bvol"]
}).plot(title="ETH Deribit volIndex vs Binance BVOL 1H")

ax.update_layout(
    yaxis_title="Volatility Index",
    xaxis_title="Date",
    xaxis_rangeslider_visible=False
)
```
